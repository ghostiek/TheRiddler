# Stuck on an Island filled with liars and truthers

We're stranded on an island and its filled with liars and [truthers](https://www.youtube.com/watch?v=IkXKP7Qodv0). Click [here](https://fivethirtyeight.com/features/dont-trust-anyone-older-than-l/) for the full article

Solution is pretty long, can be easy to get lost.

## The claims

| Islander ID        | Claim #1           | Claim #2  |
| :-------------: |:-------------:| :-----:|
|A| $B > 20$ | $D > 16$ |
|B| $C > 18$|$E < 20$|
|C|$D < 22$ | $A = 19$ |
|D| $E \neq 17$ | $B = 20$ |
|E| $A > 21$| $C < 18$ |


## What observations we can make

- A or D is lying or both based on their statement of B.

- A and C both can't be liars. One or both of them must be correct based on their statement on D.

- B or E is lying or both based on their statement of C.

- C or E is lying or both based on their statement of A.

- B and D both can't be liars, one or both have to say the truth. Based on their statement on E.

## Condensed breakdown of what we know is true

Combination of people who can't both be liars. (Must be 1 or less liars)

- A&C
- B&D

Combination of 1 or more liars. (Must be 1 or more liars)

- A&D
- B&E
- C&E

## Proof #1

### We are going to prove C says the truth.

Let's assume C is a liar.

This means A says the truth, which means D is a liar. That entails that B says the truth. We don't know the status of E.

Nevertheless, this lets us write our equation based on our age rule!

$A,B < L \leq C,D$

Under our assumptions we believe these are true statements.

$B > 20$

$D > 16$

$C > 18$

$E < 20$


If $E < 20$ and $B > 20$, E must say the truth. According to E:

$A > 21$

$C < 18$

But $A > 21$ and $C < 18$ that makes no sense if we look at our equation.

**We have thus proven by contradiction that C says the truth.**

## Proof #2

### We are going to prove A is a liar.

Lets assume A says the truth. We then know C says the truth as well. This means D is a liar, and E is a liar. We don't
know the status of B.

Using our inequality rule.

$A, C, < L \leq D, E$

Under our assumption these statements should hold:

$B >20$

$D >16$

$D < 22$

$A = 19$

$E = 17$

$B \neq 20$

We can stop early before enumerating all the statements because this makes no sense! E can't be 17 while $A = 19$


**We have thus proven that A is a liar.**

**We now know A lies, and C says the truth, by default this also tells us E is a liar**

## Proof #3

### We are going to prove D says the truth.

Lets assume D is a liar now. This means B says the truth. We use proven info to say A, E lie and C says the truth.

Our inequality rule:

$B, C < L \leq A,D,E$

Here are the statements we believe to be true under those assumptions.


$B \leq 20$

$D \leq 16$

$C > 18$

We can stop here to see this makes no sense. As its saying C is older than 18 but D is younger oe equal to 16.


**We have thus proven that D says the truth.**

**Proven so far: A, E are liars. C,D say the truth.**

## Proof #4

### We are going to prove B is a liar.


We assume B says the truth. Using what we have proven we set up the inequality that has to hold true.

$B, C, D < L \leq A, E$

$E \neq 17$

$C \geq 18$

But we have $A = 19$ when $C > 18$. That makes no sense!

**Thus we have proven that B is a liar**

### Finding L


We set up our inequality. Which states that 

$C,D < L \leq A,B,E$

as C and D are our truthers and A,B,E our liars.

We can use the values we now know true to have a quick sanity check:

$B\leq 20$

$D\leq 16$

$C\leq18$

$E\geq20$

$D<22$

$A=19$

$E\neq17$

$B=20$

$A\leq21$

$C\geq18$

All the results are actually consistent.

We will use this info to pinpoint L now.

$L = 19$ because $A = 19$ and $C \geq18$

**We have thus proven that L is 19.**

## What did we learn from the islanders

The age when people start lying is 19.

Islander A is 19.

Islander B is 20.

Islander C is 18.

Islander D is younger than or exactly 16.

Islander E is older than or exactly 20.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.special as ss

def ticket_sim(n_tickets = 50):
    (ticket1, ticket2) = (n_tickets, n_tickets)
    while ticket1 >= 0 and ticket2 >= 0:
        (ticket1, ticket2) = (ticket1 - 1, ticket2) if np.random.uniform() < 0.5 else (ticket1, ticket2 - 1)
    return (ticket1, ticket2) == (-1, 0) or (ticket1, ticket2) == (0, -1)
```

```python
def ticket_nsims(n_sims = 10000, n_tickets = 50):
    is_equal = 0
    for i in range(n_sims):
        is_equal += ticket_sim(n_tickets)
    return is_equal/n_sims
```

```python
perc = []
for i in range(100):
    perc.append(ticket_nsims(10000, i))
plt.hist(perc)
plt.show()
```

```python
df = pd.DataFrame({"Drinks Per Coupon":np.arange(100), "Probability":perc})
```

```python
g = sns.lineplot(x = "Drinks Per Coupon", y = "Probability", data = df)
plt.show()
```

```python
g = sns.scatterplot(x = "Drinks Per Coupon", y = "Probability", data = df)
g.set_title("Probability of having an empty 2nd card\n when the first runs out")
plt.show()
```
